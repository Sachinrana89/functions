This lesson covers the basics of creating classes and objects, including instance variables and methods.

A class is a blueprint for creating objects. Attributes, methods

```python
# instance variables and method
class Dog:
    # constructor
    def __init__(self,name,age):
        self.name=name
        self.age=age

# create objects
dog1=Dog("Buddy",3)
print(dog1)
print(dog1.name)
print(dog1.age)
dog2=Dog("Lucy",4)
print(dog2.name)
print(dog2.age)
```

```python
# define a class with instance methods
class Dog:
    def __init__(self, name, age):
        self.name=name
        self.age=age
    
    def bark(self):
        print(f"{self.name} says woof and his age is {self.age}")

dog1=Dog("buddy",3)
dog2=Dog("blacky",4)
dog1.bark()
dog2.bark()
```

```python
# modelling an bank account

# define a class for back account
class BankAccount:
    def __init__(self,owner,balance=0, accNo=0):
        self.owner=owner
        self.balance=balance
        self.accNo=accNo

    def deposit(self, amount):
        self.balance+=amount
        print(f"{amount} is deposited by {self.owner}. New balance is {self.balance} and its account no. is {self.accNo}")
    
    def withdraw(self, amount):
        if amount>self.balance:
            print("insufficient funds!")
        else:
            self.balance-=amount
    
    def get_balance(self):
        return self.balance
    
# create an account

account = BankAccount("krish",5000,979899)
print(account.balance)

# call instance method
account.deposit(100)
print(account.balance)

account.withdraw(40000)
print(account.balance)
```

Inheritance in Python

Inheritance is a fundamental concept in Object-Orieted Programming (OOP) that allows a class to inherit attributes and methods from another class. This lessons covers single inheritance and multiple inheritance, demonstrating how to create and use them in Python.

```python
# Inheritance
# Parent class
class Car:
    def __init__(self,windows,seats,engineType):
        self.windows=windows
        self.seats=seats
        self.engineType=engineType
    
    def drive(self):
        print(f"the person will drive {self.engineType} car which has {self.windows} windows and {self.seats} seats")

car1=Car(4,7,"petrol")
car1.drive()

# single inheritence

class Tesla(Car):
    def __init__(self, windows, doors, engineType, is_selfdriving):
        super().__init__(windows,doors,engineType)
        self.is_selfdriving=is_selfdriving

    def selfdriving(self):
        print(f"Tesla supports self driving: {self.is_selfdriving}")

tesla1=Tesla(4,5,"electric",True)
tesla1.selfdriving()
tesla1.drive()
```

```python
# multiple inheritance
# when a class inherits from more than one base class.
# base class 1

class Animal:
    def __init__(self, name):
        self.name=name

    def speak(self):
        print("subclass must implement this method")

# base class 2
class Pet:
    def __init__(self, owner):
        self.owner=owner

# Derived class
class Dog(Animal,Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} say woof"
    
# creating an object
dog=Dog("buddy","krish")
print(dog.speak())
print(f"owner:{dog.owner}")
```
