Importig Modules in Python: Modules and Packages

In python, modules and pacakages help organize and reuse code. Here's a comprehensive guide on how to import them.

```python
from math import sqrt, pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
rst=np.array([1,2,3,4,5])
print(rst)
```

```python
from packages.math import addition, subtraction
print(addition(8,9))
print(subtraction(12,9))
```

Standard Library Overview

Python's Standard Library is a vast collection of modules and packages that come bundled with Python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library.

```python
import array
arr=array.array('i',[1,2,3,4,5])
print(arr)
```

```python
import math
print(math.sqrt(16))
print(math.pi)
```

```python
# random
import random
print(random.randint(1,10))
print(random.choice(['apple','banana','cherry']))
```

```python
# file and directory Access
import os
print(os.getcwd())
```

```python
os.mkdir('new_folder')
os.rmdir('new_folder')
```

```python
# high level operation on file
import shutil
file = open("source.txt","w")
file.write("this is a source file")
file.close()
shutil.copyfile("source.txt","destination.txt")

os.remove("source.txt")
os.remove("destination.txt")
```

```python
# data Serialization
import json
data = {
    "name": "john",
    "age":30,
    "city":"new york",
    "hobbies": ["reading", "traveling", "swimming"]
}
json_str = json.dumps(data)
print(json_str)
print(type(json_str))
parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
# csv

import csv
with open('data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name','age','city'])
    writer.writerow(['john',30,'new york'])
    writer.writerow(['sachin',30,'dehradun'])

with open('data.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# datetime
from datetime import datetime, timedelta

now = datetime.now()
print(now)

yesterday = now - timedelta(days=1)
print(yesterday)
```

```python
# time
import time
print(time.time())
time.sleep(5)
print("5 seconds have passed")
print(time.time())
```
