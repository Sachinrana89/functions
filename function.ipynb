```python
# function with single parameter
def even_odd(num):
    if num%2==0:
        return "even"
    else:
        return "odd"
    
num=int(input("enter a number:"))
rst=even_odd(num)
print(rst)
```

```python
# function with multiple parameters
def calc(num1=6, num2=4, opt="+"):
    if opt=="+":
        return num1+num2
    elif opt=="-":
        return num1-num2
    elif opt=="*":
        return num1*num2
    elif opt=="/":
        return num1/num2
    else:
        return "invalid operator"
    
num1=int(input("enter first number:"))
num2=int(input("enter second number:"))
opt=input("enter operator:")
rst=calc(num1, num2, opt)
print(rst)
```

```python
# postional arguments and keywords arguments
def pos_key(*args, **kwargs):
    for val in args:
        print(f"positional argument: {val}")
    for key, val in kwargs.items():
        print(f"keyword argument: {key}={val}")

pos_key(8,1,7,1,name="sachin",city="dehradun",age=24,grade='A')
```

```python
# Example 1: this function converts temperature between celsius and fahrenheit
def convert_temperature(temp, unit):
    """this function converts temperature between celsius and fahrenheit"""
    if unit=='C':
        return (temp * 9/5)+32
    elif unit=='F':
        return (temp - 32) * 5/9
    else:
        return None
temp=float(input("enter temperature:"))
unit=input("enter unit (C/F):")
print(convert_temperature(temp, unit))
# print(convert_temperature(temp, unit))
    
```

```python
# Example 2: Password Strength Checker
def password_strength(password):
    if len(password)<8:
        return "weak"
    if not any(char.isupper() for char in password):
        return "weak"
    if not any(char.islower() for char in password):
        return "weak"
    if not any(char.isdigit() for char in password):
        return "weak"
    if not any(char in "!@#$%^&*()_+" for char in password):
        return "weak"
    if len(password)>=12:
        return "password should be between 8 and 12 characters"
    return "strong"
password=input("enter your password:")
print(password_strength(password))
```

```python
# Example 3: Calculate the Total Cost of Items in a Shopping Cart
def calculate_total(cart):
    total=0
    for item in cart:
        total+=item["price"]*item["quantity"]
    return total

cart=[
    {"name":"laptop","price":1000,"quantity":2},
    {"name":"phone","price":500,"quantity":3},
    {"name":"headphones","price":100,"quantity":5},
    {"name":"monitor","price":300,"quantity":0}
]
# calling the function
total_cost=calculate_total(cart)
print(total_cost)
```

```python
# Example 4: Checks if a string is a palindrome
def is_palindrome(s):
    s=s.replace(" ","").lower()
    print(s)
    return s==s[::-1]
print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello world"))
```

```python
# Example 6: Calculates the factorial of a number using recursion
def factorial(n):
    if n==0:
        return 1
    else:
        s=n*factorial(n-1)
        print(s)
        return s
        
print(factorial(10))
    
```

```python
# Example 6: A function To read a file and count the frequency of each word
def count_word_frequency(file_path):
    word_count={}
    with open(file_path, 'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip(".,!?\"'()")
                word_count[word]=word_count.get(word, 0)+1
    return word_count

file_path='frequency.txt'
word_frequency=count_word_frequency(file_path)
print(word_frequency)

```

```python
# Example 7: Email validation function
import re
def is_valid_email(email):
    pattern=r'^[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z0-9-.]+$'
    return re.match(pattern, email) is not None
print(is_valid_email("test@example.com"))
print(is_valid_email("invalid-email"))
```

Lambda Functions in Python 
Lambda functions are small anonymous functions defined using the Lambda keyword. They can have
any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions

```python
# synatx
lambda arguments: expression
```

```python
# normal function
def addition(a,b):
    return a+b
print(type(addition))
print(addition(5,3))

# lambda function
additionL=lambda a,b:a+b
print(type(additionL))
print(additionL(5,3))
```

```python
# normal function
def square(x):
    return x**2
print(type(square))
print(square(5))

# lambda function
squareL=lambda x:x**2
print(type(squareL))
print(squareL(5))
```

```python
# map() - applies a function to all items in a list and returns a map object
numbers=[1,2,3,4,5]
def square(x):
    return x**2
print(square(4))

print(list(map(square, numbers)))

```

The Map() function in Python
The Map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful data in a list comprehensively.

```python
def square(x):
    return x**2
numbers=[1,2,3,4,5]
print(list(map(square, numbers)))
```

```python
# lambda function with map
numbers=[1,2,3,4,5,6,7,8]
rst=list(map(lambda x:x**2, numbers))
print(rst)
```

```python
# map multiple iterables
numbers1=[1,2,3]
numbers2=[4,5,6]
added_numbers=list(map(lambda x,y:x+y, numbers1, numbers2))
print(added_numbers)
```
